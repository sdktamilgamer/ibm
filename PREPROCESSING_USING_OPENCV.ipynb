```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
pic=cv2.imread(r"/content/Elon_Musk_2015.jpg")
plt.imshow(pic)
```

```python
pic=cv2.imread(r"/content/Elon_Musk_2015.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(pic)
```

```python
flip_pic=np.flipud(pic)
plt.imshow(flip_pic,cmap='gray')
```

```python
from scipy import ndimage
rot_pic=ndimage.rotate(pic,45)
plt.imshow(rot_pic,cmap='gray')
```

```python
gauss_pic=ndimage.gaussian_filter(pic,5)
plt.imshow(gauss_pic,cmap='gray')
pic2=255-pic
cv2_imshow(pic2)
```
