```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 

from scipy import ndimage 
```

```python
import os 
```

```python
os.listdir() 
```

```python
img = cv2.imread("dhoni.jpeg")  
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
```

```python
plt.imshow(img_gray, cmap="gray")   
```

```python
cv2.imshow("GRAYSCALE",img_gray) 
```

```python
flipped = np.flipud(img_gray)  
```

```python
plt.imshow(flipped, cmap = 'gray')  
```

```python
flip_lr = np.fliplr(img_gray)
```

```python
plt.imshow(flip_lr, cmap = 'gray')  
```

```python
rot_pic = ndimage.rotate(img_gray,45)  
```

```python
plt.imshow(rot_pic, cmap = 'gray') 
```

```python
gauss_pic = ndimage.gaussian_filter(img_gray, 5)  
```

```python
plt.imshow(gauss_pic, cmap = 'gray')   
```

```python
neg = 255 - gauss_pic 
```

```python
plt.imshow(neg, cmap = 'gray')    
```

```python
negative = 255 - img_gray 
```

```python
plt.imshow(negative, cmap = 'gray')  
```
